```python
import pandas as pd
import numpy as np
from faker import Faker
import random

faker = Faker()
num_rows = 15000

data = []
for i in range(1, num_rows + 1):
    # 20% chance to leave a cell blank
    def maybe_blank(value, blank_prob=0.1):
        return value if random.random() > blank_prob else None

    name = faker.name()
    age = np.random.randint(18, 65)
    if age is not None and 18 < age <= 25:
            distance = maybe_blank(np.round(np.random.uniform(0.8, 15.0)))
            duration = maybe_blank(np.random.randint(50, 120))
            steps = maybe_blank(np.random.randint(12000, 20000))
            heartrate = maybe_blank(np.random.randint(80, 160))
            calories = maybe_blank(np.random.randint(400, 900))
        
    elif age is not None and 25 < age <= 35:
            distance = maybe_blank(np.round(np.random.uniform(0.4, 11.0)))
            duration = maybe_blank(np.random.randint(35, 100))
            steps = maybe_blank(np.random.randint(8000, 16000))
            heartrate = maybe_blank(np.random.randint(75, 165))
            calories = maybe_blank(np.random.randint(300, 800))
        
    elif age is not None and 35 < age <= 50:
            distance = maybe_blank(np.round(np.random.uniform(0.3, 8.0)))
            duration = maybe_blank(np.random.randint(25, 80))
            steps = maybe_blank(np.random.randint(6000, 12000))
            heartrate = maybe_blank(np.random.randint(70, 170))
            calories = maybe_blank(np.random.randint(200, 650))
        
    elif age is not None and 50 < age <= 65:
            distance = maybe_blank(np.round(np.random.uniform(0.2, 5.0)))
            duration = maybe_blank(np.random.randint(10, 50))
            steps = maybe_blank(np.random.randint(1000, 5000))
            heartrate = maybe_blank(np.random.randint(60, 180))
            calories = maybe_blank(np.random.randint(100, 700))
            
    #distance = maybe_blank(np.round(np.random.uniform(0.5, 15.0)))
    #duration = maybe_blank(np.random.randint(5, 120))
    #steps = maybe_blank(np.random.randint(1000, 20000))
    #heartrate = maybe_blank(np.random.randint(60, 180))
    #calories = maybe_blank(np.random.randint(100, 900))

    data.append([name, age, distance, duration, steps, heartrate, calories])

df = pd.DataFrame(data, columns=[
    'Name', 'Age', 'Distance', 'Duration', 'Steps', 'HeartRate', 'Calories'
])

# Save to CSV
df.to_csv('fitness_data_with_blanks.csv', index=False)
```

```python
from faker import Faker
import pandas as pd
import numpy as np
import random

# Initialize Faker and define some lists
fake = Faker()
num_rows = 60000

furnishing_options = ['Furnished', 'Semi-Furnished', 'Unfurnished']
parking_options = ['Yes', 'No']
locations = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai', 'Pune', 'Ahmedabad', 'Kolkata', 'Jaipur', 'Chandigarh']

# Create synthetic dataset
data = {
    'Owner Name': [fake.name() for _ in range(num_rows)],
    'Location': [random.choice(locations) for _ in range(num_rows)],
    'Size (sq ft)': np.random.randint(300, 5000, size=num_rows),
    'Bedrooms': np.random.randint(1, 6, size=num_rows),
    'Bathrooms': np.random.randint(1, 5, size=num_rows),
    'Furnishing': [random.choice(furnishing_options) for _ in range(num_rows)],
    'Age of House (years)': np.random.randint(0, 50, size=num_rows),
    'Parking Available': [random.choice(parking_options) for _ in range(num_rows)],
}

# Add 'Price' based on some logic
base_price = (
    data['Size (sq ft)'] * 5000 +
    np.array(data['Bedrooms']) * 300000 +
    np.array(data['Bathrooms']) * 200000 -
    np.array(data['Age of House (years)']) * 10000 +
    np.array([100000 if f == 'Furnished' else 50000 if f == 'Semi-Furnished' else 20000 for f in data['Furnishing']]) +
    np.array([50000 if p == 'Yes' else 0 for p in data['Parking Available']])
)

# Add noise
noise = np.random.normal(0, 500000, num_rows)
data['Price'] = (base_price + noise).astype(int)

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV (optional)
df.to_csv("real_estate_data_60000.csv", index=False)

print("✅ Real estate dataset with 60,000 rows generated successfully!")
```

```python
from faker import Faker
import pandas as pd
import numpy as np
import random

# Initialize Faker
fake = Faker()
num_rows = 60000

# Options
furnishing_options = ['Furnished', 'Semi-Furnished', 'Unfurnished']
parking_options = ['Yes', 'No']
locations = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai', 'Pune', 'Ahmedabad', 'Kolkata', 'Jaipur', 'Chandigarh']
property_types = ['Apartment', 'Villa', 'Independent House', 'Studio', 'Penthouse']
nearby_facilities = ['Yes', 'No']

# Generate synthetic data
data = {
    'Owner Name': [fake.name() for _ in range(num_rows)],
    'Location': [random.choice(locations) for _ in range(num_rows)],
    'Size (sq ft)': np.random.randint(300, 5000, size=num_rows),
    'Bedrooms': np.random.randint(1, 6, size=num_rows),
    'Bathrooms': np.random.randint(1, 5, size=num_rows),
    'Furnishing': [random.choice(furnishing_options) for _ in range(num_rows)],
    'Age of House (years)': np.random.randint(0, 50, size=num_rows),
    'Parking Available': [random.choice(parking_options) for _ in range(num_rows)],
    'Type of Property': [random.choice(property_types) for _ in range(num_rows)],
    'Nearby Facilities': [random.choice(nearby_facilities) for _ in range(num_rows)],
}

# Pricing multipliers
furnish_bonus = {'Furnished': 100000, 'Semi-Furnished': 50000, 'Unfurnished': 20000}
parking_bonus = {'Yes': 50000, 'No': 0}
property_bonus = {
    'Apartment': 0,
    'Villa': 1000000,
    'Independent House': 700000,
    'Studio': -100000,
    'Penthouse': 1200000
}
facility_bonus = {'Yes': 50000, 'No': 0}

# Compute price
base_price = (
    data['Size (sq ft)'] * 5000 +
    np.array(data['Bedrooms']) * 300000 +
    np.array(data['Bathrooms']) * 200000 -
    np.array(data['Age of House (years)']) * 10000 +
    np.array([furnish_bonus[f] for f in data['Furnishing']]) +
    np.array([parking_bonus[p] for p in data['Parking Available']]) +
    np.array([property_bonus[t] for t in data['Type of Property']]) +
    np.array([facility_bonus[f] for f in data['Nearby Facilities']])
)

# Add random noise
noise = np.random.normal(0, 500000, num_rows)
data['Price'] = (base_price + noise).astype(int)

# Final DataFrame
df = pd.DataFrame(data)
df.to_csv("real_estate_data_final.csv", index=False)

print("✅ Dataset generated with updated price logic including 'Type of Property' and 'Nearby Facilities'")
```
